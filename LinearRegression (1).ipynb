```python
import pandas
from pandas import DataFrame
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pandas.read_csv('cost_revenue_clean (1).csv')
```

```python
data
```

```python
data.describe()
```

```python
X = DataFrame(data,columns = ['production_budget_usd'])
Y = DataFrame(data,columns = ['worldwide_gross_usd'])
```

```python
plt.figure(figsize = (10,6))
plt.scatter(X,Y,alpha=0.3)
plt.title("Film Cost vs Global Revenue")
plt.xlabel('production budget')
plt.ylabel('worldwide gross')
plt.ylim(0,3000000000)
plt.xlim(0,450000000)
plt.show()
```

```python
regression = LinearRegression()
regression.fit(X,Y)
```

Slope Coeffficient

```python
regression.coef_  #alpha1
```

```python
#Intercept
regression.intercept_
```

```python
plt.figure(figsize = (10,6))#For the size of graph
plt.scatter(X,Y,alpha=0.3)#For the ligthning of vdata
plt.plot(X,regression.predict(X),color = 'red',linewidth=4)
#In above code we have used our x values and x predicted values to plot a linear h()

plt.title("Film Cost vs Global Revenue")
plt.xlabel('production budget')
plt.ylabel('worldwide gross')
plt.ylim(0,3000000000)
plt.xlim(0,450000000)
plt.show()
```

```python
regression.score(X,Y)
```
